# RMSE Comparison (Basic MF, Social MF, SolRec)

```python
def change_width(ax, new_value) :
    for patch in ax.patches :
        current_width = patch.get_width()
        diff = current_width - new_value

        # we change the bar width
        patch.set_width(new_value)

        # we recenter the bar
        patch.set_x(patch.get_x() + diff * .5)
```

```python
rmsedata = [1.1796581077410584, 1.3267437437614205, 1.1659027340426056]
ax= sns.barplot(['Basic MF', 'SocialMF', 'SoRec'], rmsedata, linewidth=9, palette='rainbow')
change_width(ax, .35)
plt.title("RMSE Comparison")
plt.ylabel('RMSE')
plt.xlabel('Algorithms')
```
